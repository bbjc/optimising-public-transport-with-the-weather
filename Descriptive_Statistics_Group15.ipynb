<b>On regroupe par jour fériés<b>

```python
import holidays
# Récupérer les jours fériés en France
fr_holidays = holidays.France(years=df_freq_agg['date'].dt.year.unique())

# Ajouter une colonne 'jour_ferie' (True si c'est un jour férié, sinon False)
df_freq_agg['jour_ferie'] = df_freq_agg['date'].isin(fr_holidays)

df_freq_agg
```

```python

# Regrouper par jour férié et calculer la somme des fréquentations
df_grouped = df_freq_agg.groupby('jour_ferie')['Frequentation'].sum().reset_index()

# Remplacer les booléens par des chaînes pour l'affichage
df_grouped['jour_ferie'] = df_grouped['jour_ferie'].replace({True: 'Jour Férié', False: 'Jour Normal'})

# Tracer le graphique à barres
plt.figure(figsize=(8, 5))
plt.bar(df_grouped['jour_ferie'], df_grouped['Frequentation'], color=['#ff7f0e', '#1f77b4'])
plt.title('Répartition des Fréquentations par Type de Jour')
plt.xlabel('Type de Jour')
plt.ylabel('Fréquentation Totale')
plt.grid(axis='y')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

La nombre de jour férié est très faible, on décide de ne pas les utiliser dans notre étude

<b>On regroupe par saisons<b>

```python
# Fonction pour attribuer une saison en fonction du mois 
def assign_season(month):
    if month in ['Décembre', 'Janvier', 'Fevrier']:
        return 'Hiver'
    elif month in ['Mars', 'Avril', 'Mai']:
        return 'Printemps'
    elif month in ['Juin', 'Juillet', 'Août']:
        return 'Été'
    elif month in ['Septembre', 'Octobre', 'Novembre']:
        return 'Automne'

# Ajouter une colonne 'Saison' basée sur le mois
df_freq_agg['Saison'] = df_freq_agg['mois'].apply(assign_season)

df_freq_agg
```

```python
# Regrouper par saison et calculer la somme des fréquentations
df_groupedsaison = df_freq_agg.groupby('Saison')['Frequentation'].mean().reset_index()
df_groupedsaison.dtypes
```

```python
# Créer un graphique à barres
plt.bar(df_groupedsaison['Saison'], df_groupedsaison['Frequentation'], color=['#4CAF50', '#2196F3', '#FF5722', '#FFC107'])

# Ajouter des titres et des étiquettes
plt.title('Fréquentations par Saison')
plt.xlabel('Saison')
plt.ylabel('Moyenne de Fréquentations')

# Afficher le graphique
plt.show()
```

La saison d'été présente des fréquentations plus faible, cela est logique, elle présente beaucoup de vacances

On peut supposer que la fréquentation est plus importante en automne à cause de la rentré des classes

<b>On regroupe par semaine<b>

```python
moyenne_frequentation = df_freq_agg.groupby('semaine')['Frequentation'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='semaine', y='Frequentation', data=moyenne_frequentation)
plt.title('Fréquentation moyenne pour le métro a par mois')
plt.xlabel('Semaine')
plt.ylabel('Fréquentation moyenne')
plt.tight_layout()
plt.show()
```

On retrouve une baisse de fréquentation lors des semaines de vacances

<b>On regroupe par jour de la semaine<b>

```python
# Regrouper par jour et calculer la somme des fréquentations
df_grouped = df_freq_agg.groupby('jourSemaine')['Frequentation'].sum().reset_index()

ordre_jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']

df_grouped['jourSemaine'] = pd.Categorical(df_grouped['jourSemaine'], categories=ordre_jours, ordered=True)

df_grouped = df_grouped.groupby('jourSemaine')['Frequentation'].sum().reset_index()

# Tracer le graphique à barres
plt.figure(figsize=(10, 6))
plt.bar(df_grouped['jourSemaine'], df_grouped['Frequentation'], color='skyblue')
plt.title('Répartition des Fréquentations par Jour')
plt.xlabel('Jour de la Semaine')
plt.ylabel('Fréquentation Totale')
plt.xticks(rotation=45)  # Rotation des labels sur l'axe des x pour une meilleure lisibilité
plt.grid(axis='y')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

La fréquentation est maximale du mardi au vendredi
Elle baisse fortement le week-end, surtout le dimanche

<h2>Exploration du Dataset join<h2>

On s'intéresse ici seulelemnt au mois de novembre pour plus de simplicité

<b>On importe les deux datatset<b>

```python
df_freq = pd.read_csv('star_frequentation_nov.csv', sep=';')
df_meteo = pd.read_csv('meteo_nov.csv', sep=';')
```

<b>On transforme les données de fréquentation<b>

On arrondie les données de frequentation

```python
df_freq['Frequentation'] =  df_freq['Frequentation'].str.split(',').str[0].astype('int')
```

On somme la frequentation par ligne et par jour

```python
df_freq['heure'] = df_freq['TrancheHoraire15mn'].str.slice(start=0, stop=5)

df_freq['heure'] = pd.to_datetime(df_freq['heure'], format='%H:%M', errors='coerce')

df_freq['heure'] = df_freq['heure'].dt.round('H')

df_freq['heure'] =df_freq['heure'].dt.strftime('%H:%M')

df_freq['DateFreq'] = pd.to_datetime(df_freq['DateFreq'],format='%d/%m/%Y')

df_freq['date'] = df_freq['DateFreq'].dt.strftime('%Y-%m-%d') + 'T' +  df_freq['heure']

new_df_freq = df_freq.groupby(['date', 'identifiantLigne', 'NomCourtLigne'], as_index=False).agg({
    'Frequentation': 'sum'
})

new_df_freq
```

<b>On transforme les données météo<b>

On récupère les variables voulues dans le dataset de la météo

```python
df_meteo = df_meteo[['Date', 'Température (°C)', 'Précipitations dans les 3 dernières heures', 'Rafale sur les 10 dernières minutes']]
df_meteo
```

On transforme les date en type datetime

```python
df_meteo['date'] = df_meteo['Date'].str.slice(0,16)
df_meteo = df_meteo.drop(columns=['Date'])
```

<b>On join nos base de données<b>

```python
df = df_meteo.merge(new_df_freq, on='date', how='left')
```

On transforme notre variable de date en deux variables heure et date(jour)

```python
df['datetime'] = df['date']
df['datetime'] = pd.to_datetime(df['datetime'])


df['heure'] = df['date'].str.split('T').str[1]
df['date'] = df['date'].str.split('T').str[0]

df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
```

<b>On calcul la correlation<b>

<b>Sur toute les lignes<b>

```python
the_numerical_columns = ['Température (°C)', 'Précipitations dans les 3 dernières heures',
                     'Rafale sur les 10 dernières minutes', 'Frequentation']


plt.figure(figsize=(8, 6))
sns.heatmap(df[the_numerical_columns].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap des corrélations entre les variables")
plt.show()
```

La corrélation est très faible lorsque l'on aggrége les fréquentation sur tous le réseau

<b>Seulement sur la ligne de métro<b>

```python
the_numerical_columns = ['Température (°C)', 'Précipitations dans les 3 dernières heures',
                     'Rafale sur les 10 dernières minutes', 'Frequentation']


plt.figure(figsize=(8, 6))
sns.heatmap(df_metro[the_numerical_columns].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap des corrélations entre les variables")
plt.show()
```
